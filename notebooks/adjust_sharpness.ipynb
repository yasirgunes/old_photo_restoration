```python
from utils.histogrameq_contrast_brightness import equalize_histogram, adjust_brightness_contrast
from utils.mask_inpaint import inpaint_scratches
from utils.adjust_sharpness import unsharp_masking
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_two_images(image1, image2, title1, title2):
    plt.figure(figsize=(12, 6))

    # Original image
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(image1)
    plt.axis('off')

    # CLAHE equalized image
    plt.subplot(1, 2, 2)
    plt.title("Result Image")
    plt.imshow(image2)
    plt.axis('off')

    plt.tight_layout()
    plt.show()
        
```

```python
# read the image
original_image = cv2.imread('image.png')
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

# equalize the histogram
hist_eq_image = equalize_histogram(original_image)

# display the original image and the equalized image
display_two_images(original_image_rgb, hist_eq_image, 'Original Image', 'Equalized Image')

# adjust the brightness and contrast
adjusted_image = adjust_brightness_contrast(hist_eq_image)

# display the original image and the adjusted image
display_two_images(hist_eq_image, adjusted_image, 'Original Image', 'Adjusted Image')
```

```python
# Apply unsharp masking
sharpened_image = unsharp_masking(adjusted_image)

# display the original image and the sharpened image
display_two_images(adjusted_image, sharpened_image, 'Original Image', 'Sharpened Image')
```
